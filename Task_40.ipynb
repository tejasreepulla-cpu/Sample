```python
data={'first name':['Arun','jebu','venkat','rekha','majid','mohsin'],'last name':['kumar','jacob','raghavan','singh','khan','khan'],'employmenttype':['service','business','student','service','business','business'],'country':['india','usa','usa','sweden','australia','germany']}
df=pd.DataFrame(data,columns=['first name','last name','employmenttype','country'])
print(df)
```

```python
df1=pd.get_dummies(df['employmenttype'],dtype='int')
df2=pd.get_dummies(df['country'],dtype='int')
print(df1)
print(df2)
```
